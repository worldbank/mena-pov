```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
from selenium.webdriver.chrome.options import Options
```

```python
# Specify the base download directory
base_download_directory = "/Users/chitrab/Downloads/Quarterly"

# Specify the path to the ChromeDriver executable
chromedriver_path = "/Users/chitrab/chromedriver_mac64/chromedriver"

# Define the range of t_no values
start_t_no = 13
end_t_no = 70

# Define the range of year1 values
start_year1 = 2007
end_year1 = 2023

# Define the range of round1 values
start_round1 = 1
end_round1 = 4

# Set the download options for ChromeDriver
chrome_options = webdriver.ChromeOptions()
prefs = {
    "download.default_directory": base_download_directory,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
}
chrome_options.add_experimental_option("prefs", prefs)

# Create a ChromeDriver instance with the specified options
driver = webdriver.Chrome(chromedriver_path, options=chrome_options)

# Loop through round1 values
for round1 in range(start_round1, end_round1 + 1):
    # Create a round folder
    round_folder_path = os.path.join(base_download_directory, f"Round {round1}")
    if not os.path.exists(round_folder_path):
        os.makedirs(round_folder_path)

    # Loop through t_no values
    for t_no in range(start_t_no, end_t_no + 1):
        # Loop through year1 values
        for year1 in range(start_year1, end_year1 + 1):
            try:
                # Construct the URL with the current t_no, year1, and round1 values
                url = f"http://www.dos.gov.jo/owa-user/owa/emp_unemp.show_tables1?lang=E&year1={year1}&round={round1}&t_no={t_no}"

                # Open the URL
                driver.get(url)

                # Wait for the page to load (you may need to adjust the sleep duration based on the page load time)
                time.sleep(2)

                # Find the <img> element within the specified XPath
                img_element = driver.find_element(By.XPATH, '/html/body/center[3]/table/tbody/tr/td[3]/a/img')

                # Click on the img element
                img_element.click()

                # Wait for the file to be downloaded (you may need to adjust the sleep duration based on the file download time)
                time.sleep(5)

                # Move the downloaded file to the respective year folder within the round folder
                original_file_path = os.path.join(base_download_directory, f"Table {t_no}.xls")
                year_folder_path = os.path.join(round_folder_path, str(year1))
                if not os.path.exists(year_folder_path):
                    os.makedirs(year_folder_path)
                new_file_path = os.path.join(year_folder_path, f"Table {t_no}.xls")
                os.rename(original_file_path, new_file_path)

            except Exception as e:
                print(f"Error occurred while downloading Table {t_no} for year {year1} and round {round1}: {str(e)}")
                continue

# Close the web driver
driver.quit()
```
